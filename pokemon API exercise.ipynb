```python
import requests, json

class Pokedek:
    pokedek = []
    bul_data=None
    
    @classmethod
    def add(cls, pokemon_data):
        cls.pokedek.append(pokemon_data)
                 
    @classmethod
    def Get_data(cls, pokemon_name):
        response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}')
        data=response.json()
        name=f"{data['types'][0]['type']['name']}"
        p_dict = {
            'abilities': data['abilities'][0]['ability']['name'],
            'name': data['name'],
            'type':name,
            'height':data['height'],
            'weight':data['weight']
        }
        cls.add(p_dict)
    
    @classmethod
    def show(cls):
        for p in cls.pokedek:
            print(p)
            
    @classmethod
    def sort_by(cls):
        pokemon_dict = {}
        for poke in cls.pokedek:
#             pk_dict = {}
#             print(poke)
            if poke['type'] not in pokemon_dict:
                pokemon_dict[poke['type']]={}
#             print(pokemon_dict)
            pokemon_dict[poke['type']].update({
                poke['name']:{
                    'abilities': poke['abilities'],
                    'height':poke['height'],
                    'weight':poke['weight']
                }
            })
        return pokemon_dict
            

        
Pokedek.Get_data('rattata')
Pokedek.Get_data('bulbasaur')
Pokedek.Get_data('raticate')
Pokedek.Get_data('charmander')
Pokedek.Get_data('squirtle')
Pokedek.Get_data('wartortle')
Pokedek.Get_data('charizard')
Pokedek.Get_data('pikachu')
Pokedek.Get_data('raichu')
Pokedek.Get_data('diglett')
Pokedek.Get_data('Sharpedo')
Pokedek.Get_data('Ponyta')
Pokedek.Get_data('Rapidash')
Pokedek.Get_data('Seel')
Pokedek.Get_data('Dewgong')
Pokedek.Get_data('Wailmer')
Pokedek.Get_data('Wailord')
Pokedek.Get_data('Stunky')
Pokedek.Get_data('Sealeo')
Pokedek.Get_data('Togekiss')

Pokedek.sort_by()
```
